## **Analisis de componentes principales**



Geometria de variables

La **media** de una variable puede ser representada en la recta real.

En física, la representación es el punto de equilibrio, o en otras palabras, el centro de gravedad.


En la mayoria de los casos se quiere saber las coordenadas respecto al centro de gravedad, las distancias.


La varianza representa el momento de incercia de los 10 puntos respecto al centro de gravedad.


Nota: Centrado: restar el valor de la media con los valores para obtener las coordenas en la recta real.


Para obtener un diagrama de dispersion de datos centrados, se debe poner el eje en el valor de la media.

1.png

Al final se calcula el eje de mayor proyeccion y se rotan los ejes respecto a los puntos.

En ese sistema de ejes, existe correlacion entre las dos variables, y da la misma varianza, esa es la ventaja.

Ideal para el análisis multivariado.
